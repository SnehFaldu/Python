**Post Lab - 1**

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000  # Sampling frequency
t = np.linspace(0, 1, fs, endpoint=False)

sine_5 = np.sin(2 * np.pi * 5 * t)
sine_10 = np.sin(2 * np.pi * 10 * t)

combined = sine_5 + sine_10

plt.figure(figsize=(10,5))
plt.plot(t, combined, label="5 Hz + 10 Hz Sine")
plt.title("Addition of Two Sine Waves (5 Hz and 10 Hz)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

**Post Lab - 2**

```python
fs = 500
t = np.linspace(0, 2, 2*fs, endpoint=False)

sine_5 = np.sin(2 * np.pi * 5 * t)
cosine_10 = np.cos(2 * np.pi * 10 * t)

product = sine_5 * cosine_10

plt.figure(figsize=(10,5))
plt.plot(t, product, label="5 Hz Sine  10 Hz Cosine")
plt.title("Multiplication of Sine and Cosine Signals")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

**Post Lab - 3**

```python
fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)

sine = np.sin(2 * np.pi * 5 * t)
shifted = np.sin(2 * np.pi * 5 * (t - 0.1))  # time-shift by 0.1s

plt.figure(figsize=(10,5))
plt.plot(t, sine, label="Original 5 Hz Sine")
plt.plot(t, shifted, '--', label="Shifted by 0.1s")
plt.title("Time Shift of Sine Wave")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

**Post Lab - 4**

```python
fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)

sine = np.sin(2 * np.pi * 10 * t)
scaled = 3 * sine

plt.figure(figsize=(10,5))
plt.plot(t, sine, label="Original 10 Hz Sine")
plt.plot(t, scaled, '--', label="Scaled by 3")
plt.title("Amplitude Scaling of Sine Wave")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

**Post Lab - 5**

```python
fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)

sine = np.sin(2 * np.pi * 5 * t)
reversed_signal = sine[::-1]  # reverse the array

plt.figure(figsize=(10,5))
plt.plot(t, sine, label="Original 5 Hz Sine")
plt.plot(t, reversed_signal, '--', label="Reversed Sine")
plt.title("Time Reversal of Sine Wave")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```
